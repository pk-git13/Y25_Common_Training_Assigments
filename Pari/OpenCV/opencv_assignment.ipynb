## 3. Basic Image Functions
**Task:**
1. Load `park.jpg`.
2. Convert it to grayscale.
3. Apply Gaussian Blur with a $(7,7)$ kernel.
4. Perform Canny Edge detection on the blurred image.


```python
# Write your code here
image2=cv.imread("open_cv files/park.jpg")
image2_rgb = cv.cvtColor(image2, cv.COLOR_BGR2RGB)

gray=cv.cvtColor(image2,cv.COLOR_RGB2GRAY)




blur = cv.GaussianBlur(gray, (7,7), 0)
blur_rgb = cv.cvtColor(blur, cv.COLOR_BGR2RGB)

canny = cv.Canny(blur, 125, 175)
canny_rgb = cv.cvtColor(canny, cv.COLOR_BGR2RGB)




plt.imshow(image2_rgb)
plt.title("priginal")
plt.axis("off")
plt.show()

plt.imshow(gray,cmap='gray')
plt.title("gray")
plt.axis("off")
plt.show()

plt.imshow(blur_rgb)
plt.title("Blurred Image")
plt.axis("off")
plt.show()

plt.imshow(canny_rgb)
plt.title("canny")
plt.axis("off")
plt.show()

```

## 4. Image Transformations
**Task:**
1. Create a function `translate(img, x, y)` to shift an image.
2. Create a function `rotate(img, angle, rotPoint=None)` to rotate an image.
3. Apply a flip transformation to `park.jpg`.

```python
# Write your code here
def translate(img,x,y):
 
    M = np.float32([
        [1, 0, x],
        [0, 1, y]
    ])
    shifted = cv.warpAffine(img, M, (img.shape[1], img.shape[0])) #warpAffine wants input dimensions in width,height and not height,width
    
    shifted_rgb = cv.cvtColor(shifted, cv.COLOR_BGR2RGB)

    plt.imshow(shifted_rgb)
    plt.axis("off")
    plt.show()

def rotate(img,angle,rotPoint=None):
    height,width=img.shape[:2]
    
    R=cv.getRotationMatrix2D((width//2,height//2),angle,1.0)
    rotated=cv.warpAffine(img,R,(img.shape[1],img.shape[0]))

    rotated_rgb = cv.cvtColor(rotated, cv.COLOR_BGR2RGB)

    plt.imshow(rotated_rgb)
    plt.axis("off")
    plt.show()

translate(image2,30,-50)
rotate(image2,45)

vertical_flip=cv.flip(image2,0)#0 for vertical flip,1 for horizontal flip and -1 for both
vertical_flip_rgb = cv.cvtColor(vertical_flip, cv.COLOR_BGR2RGB)

plt.imshow(vertical_flip_rgb)
plt.axis("off")
plt.show()



    
```

## 5. Contour Detection
**Task:**
1. Load `cats.jpg`.
2. Convert to grayscale and apply Gaussian blur.
3. Find the edges using Canny.
4. Use `cv.findContours` to find all contours.
5. Print the count and draw them on a blank image.

```python
# Write your code here
gray_1 = cv.cvtColor(img, cv.COLOR_BGR2GRAY)


blur_1 = cv.GaussianBlur(gray_1, (7,7), 0)

canny_1= cv.Canny(blur_1, 125, 175)

contours, hierarchies = cv.findContours(
    canny_1,
    cv.RETR_LIST,
    cv.CHAIN_APPROX_SIMPLE
)
print(len(contours))

blank=np.zeros((500,500,3))
cv.drawContours(blank,contours,-1,(255,255,255))#-1 for drawing all contours


plt.imshow(blank)
plt.axis("off")
plt.show()
```

## 8. Bitwise Operations
**Task:**
1. Create a $400 \times 400$ blank image.
2. Create a rectangle and a circle on separate blank images.
3. Perform Bitwise AND, OR, XOR, and NOT.

```python
# Write your code here
blank_r=np.zeros((400,400,3))
blank_c=np.zeros((400,400,3))

rectangle=cv.rectangle(blank_r,(170,190),(230,210),(255,255,255))
circle=cv.circle(blank_c,(200,200),20,(255,255,255))

bitwise_and=cv.bitwise_and(rectangle,circle)
bitwise_or=cv.bitwise_or(rectangle,circle)
bitwise_not_rectangle=cv.bitwise_not(rectangle)
bitwise_not_circle=cv.bitwise_not(circle)
bitwise_xor=cv.bitwise_xor(rectangle,circle)



plt.imshow(bitwise_and)
plt.axis("off")
plt.show()



plt.imshow(bitwise_or)
plt.axis("off")
plt.show()



plt.imshow(bitwise_not_rectangle)
plt.axis("off")
plt.show()

plt.imshow(bitwise_not_circle)
plt.axis("off")
plt.show()


plt.imshow(bitwise_xor)
plt.axis("off")
plt.show()
```

## 9. Thresholding
**Task:**
1. Load `cats.jpg` and convert to grayscale.
2. Apply Simple Thresholding (Binary and Inverse).

```python
# Write your code here
img=cv.imread("open_cv files\cat.jpg")


_, binary = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
_, binary_inv = cv.threshold(img, 127, 255, cv.THRESH_BINARY_INV)

binary_rgb = cv.cvtColor(binary, cv.COLOR_BGR2RGB)
plt.imshow(binary_rgb)
plt.axis("off")
plt.show()

binary_inv_rgb = cv.cvtColor(binary_inv, cv.COLOR_BGR2RGB)
plt.imshow(binary_inv_rgb)
plt.axis("off")
plt.show()

```
